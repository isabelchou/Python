```python
# Show messages to customers and users:

def greet():
    print('Welcome: ') 

    
# Goodbye message to users:
def bye():
    print("Thank you for your purchase. Goodbye!") 

    
# Show admin functions:
def adm_func():
    print('\n a reset product quantity \n b reset quantity of coins stored \n c enter user screen (vending system starts)')

    
# Show user functions
def user_func():
    print ('\n a select products from product list \n b show transaction records \n c admin use: machine not working (show error message)')

    
# When needing to separate content to make screen more readable for users:
def skipline():  
    print('\n')

    
```

```python
# Add different products and define attributes: product number, name, cost and available quantity
class Products:    
    def __init__(self, num, name, cost, availty = 10):  
        self.num = num
        self.name = name
        self.cost = cost
        self.availty = availty
        
    
    # Show list of available products
    def show_avail(listofprod):   
        print("\nAvailable products: ")
        for prod in listofprod:
            if prod.availty > 0:                   
                print(f'{prod.num}. {prod.name}   Cost: {prod.cost / 100} dollars;   {prod.availty} Available ')
        # Give information about unavailable products:
        skipline()    
        for prod in listofprod:
            if prod.availty == 0:                   
                print(f'{prod.name} is currently unavailable \n')  
        
        
    # Admin function: reset product quantity
    def res_quan(productlist):
        res_prod = True 
        while res_prod:        
            Products.show_avail(productlist)
            try:
                for_prodnum = int(input('Please enter product number to reset: '))                  
                if for_prodnum not in range(1,7):
                    print('Sorry, choice does not match any product number.Please enter number 1-6')
                    continue
                else:
                    change_to = int(input('Change availability to: '))
                    for prod in productlist:
                        if prod.num == for_prodnum:
                            prod.availty = change_to
                    res_prod = False
                    break
            # Warning given when input other than numbers entered
            except ValueError:
                    print("Sorry, number expected. Please start again.")
                    continue   
    
    
    # User selects items durng purchase
    def choose_item(productlist, sugarstock):     
        Products.show_avail(productlist) 
        select_prod = True
        while select_prod:
            prod_choice = input("Enter number 1-6 to select product, select 'cancel' to return to homescreen: ")
            # User can cancel choice and go back to home screen
            if prod_choice == 'cancel':                    
                return 'cancel'
            # Warning given when input is not digit, or does not match product number
            elif not prod_choice.isdigit() or int(prod_choice) not in range(1, 7):
                print("\n Sorry, please enter number 1-6 or select 'cancel' \n")
                continue
            else:                 
                itm = productlist[int(prod_choice) - 1]
                if itm.availty == 0:
                    print('Sorry, product is currently unavailable.')
                    continue
                else:
                    # Reminds user when sugar is out of stock
                    if itm == coffee and sugarstock == 0:
                        print('Please note: sugar is unavailable for coffee.')
                    else:
                        pass
                        # Show unit price and ask user whether to continue purchase
                        print(f"Unit price for {itm.name} is {itm.cost / 100} dollars.")
                        choose = True
                        while choose:
                            ctn = input("Select 'continue' to confirm, 'restart' to select another item, 'cancel' to cancel order: ")                        
                            # User selects restart to reset item (transaction history will be recorded when items confirmed and payment completed)
                            if ctn == 'restart':
                                return 'restart'                           
                            elif ctn == 'cancel':
                                return 'cancel'
                            elif ctn == 'continue':
                                return itm
                            else:
                                print('Sorry, option not recognised.')
                            
                            

    def set_qtt(self):
        while True:
            how_many_str = input('Please enter quantity needed: ')
            # Reminder given when input is not a number
            if not how_many_str.isdigit():
                print('Sorry, input not recognised. Please enter a number.')
                continue
            else:
                how_many = int(how_many_str)                
                if how_many > self.availty:
                    print(f'Sorry, only {self.availty} {self.name} are available.')
                    # Reminds user when selected quantity exceeds availablity and allows user to reset quantity or cancel item.
                    re_ent = input("'continue' to re-enter quantity, any other key to cancel item: ")
                    if re_ent == 'continue':
                        continue
                    else:
                        res_can = input("Press 'restart' to select another item, 'cancel' to return to home screen: ")
                        if res_can == 'restart':
                            return 0
                        elif res_can == 'cancel':
                            return -1
                        elif res_can != 'restart' and res_can != 'cancel':
                            print("Sorry, input not recognised. Back to home screen.")
                            return -1 
                elif not how_many > 0:
                    print(f'Sorry, please enter valid quantity.')
                    re_ent = input("'continue' to re-enter quantity, any other key to cancel item: ")
                    if re_ent == 'continue':
                        continue
                    else:
                        cancelchoice = True
                        while cancelchoice:
                            res_can = input("Press 'restart' to select another item, 'cancel' to return to home screen: ")
                            if res_can == 'restart':
                                return 0
                            elif res_can == 'cancel':
                                return -1
                            elif res_can != 'restart' and res_can != 'cancel':
                                print("Sorry, input not recognised. Back to home screen")
                                return -1
                # Show total price of selected quantity of products and confirm whether quantity needs to be reset
                else:                                                                            
                    print(f'Price for {how_many} {self.name} is {how_many * self.cost / 100} dollars \n')
                    res = True
                    while res:
                        res_can = input("Select 'continue' to confirm, 'restart' to reset quantity: ")
                        if res_can == 'restart':
                            res = False
                            continue
                        elif res_can == 'continue':
                            return how_many 
                        else:
                            print('Sorry, choice not recognised.')
                            continue
                    

    def dispense(product, quantity):
        return product, quantity 
        # Machine will dispense the stated quantity of items
        

    def prepcoffee(sugarstock, cup_num):    
        if sugarstock > 0: 
            sugar = input(f"'Y' to add sugar to your coffee, any other key to skip sugar: ")                               
            if sugar.lower() == 'y':                                                                                                                                    
                manual = input("Enter 1 to add sugar automatically, any other key to mix sugar by hand: ")
                if manual == '1':
                    print(f'Please wait {3 * cup_num} seconds for sugar preparation.') 
                    # 3 seconds needed for sugar to be mixed automatically for each cup of coffee
                else:
                    print(f'Please collect sugar.')                     
                if sugarstock < cup_num:
                    sugarstock = 0
                else:
                    sugarstock -= cup_num 
            else:
                pass                     
        else:
            print("Please note: sugar is not currently available.")   
            # Alert message given again when sugar is out of stock                                  
        print(f'Please wait for {30 * cup_num} seconds for water to be boiled.')  
        # Displays wait time for water boiling
        return sugarstock
    
```

```python
# System accepts different types of coins
class Coin:
    def __init__(self, name, val, stor = 50):
        self.name = name
        self.val = val
        self.stor = stor
        # Coin attributes: name, value and storage
      
    
    # Admin function: show list of coins stored and quantity stored for each
    def show(listofcoins = []): 
        for coin in listofcoins:
            print(f'{coin.name}      quantity stored: {coin.stor}')  
    
    
    # For given amount of refund required, use available coin to make up the refund.  
    def get_ref(to_ref, coinlist):
        while to_ref >= min(coin.val for coin in coinlist if coin.stor > 0):
            for cointype in coinlist:
                if to_ref >= cointype.val and cointype.stor > 0:
                    cointype.stor -= 1
                    to_ref -= cointype.val
                    continue
                else:
                    continue
        # If stored coins cannot make the total refund, the remaning amount is recorded. Later in vending process the remaining amount wll be added to user credit 
        return to_ref
    
    
    # Admin function: reset stored number of coins
    def reset(listofcoins):
        skipline()
        res_coin = True
        while res_coin:        
            Coin.show(listofcoins)  
            try:
                for_coinval = int(input('Reset for coin (enter value in cents, e.g 50, 100($1), 200($2): '))                      
                # Exception handling: input value outside stored coin range
                if for_coinval not in [10, 20, 50, 100, 200]:
                    print('Sorry, system only stores coins of 10c, 20c, 50c, $1 and $2.')
                    try_again = input("Select'continue' to enter another value, or any other key to return to menu: ")
                    if try_again == 'continue':
                        continue
                    else:
                        res_coin = False
                        break
                else:
                    change_to = int(input('Number of coin stored changes to: '))
                    for coin in listofcoins:
                        if coin.val == for_coinval:
                            coin.stor = change_to
                    res_coin = False
                    break
            except ValueError:
                    print("Sorry, number expected. Please start again.")
                    continue 
                    
```

```python
# Add admin procedures
class Admin:
    def __init__(self, password):
        self.password = password

        
    # During first installation, system will call this function for admin staff to set password   
    def set_pswd():    
        while True:
            pswd = input('First installation. Please set admin password: ')
            if len(pswd) == 0:
                print('Sorry, password cannot be empty')
                continue
            else:            
                return pswd 

    
     # When machine not working, staff can use admin password set error status message 
    def err_msg(password):
        # Show error message (admin function)
        ver_pswd = True
        while ver_pswd:
            ent_pswd = input('\nPlease enter admin password (case sensitive): ')
            if ent_pswd == password:                    
                verified = True                    
                while verified:
                    print ("\n \n \n Machine out of order. \n \n \n")
                    # When problem solved, staff can use admin password to cancel error message
                    solved = input('(If problem solved, enter admin password to cancel this message: )')
                    if solved == password: 
                        skipline()
                        verified = False
                        ver_pswd = False
                        break
                    else:
                        print('Sorry, password incorrect.')
                        continue
            else:
                # Warning given when password is incorrect
                print("Sorry, password does not match our record.")
                back_menu = input("Press 'continue' to try again, any other key to go back to main menu: ")
                if back_menu == 'continue':
                    continue
                else:
                    ver_pswd = False
                    break
                    
```

```python
# Each transaction has attributes: transaction number, user, date, payment and items purchased
class Transaction:
    def __init__(self, trans_no, accountname, date, pay, shoplist = []):
        self.trans_no = trans_no
        self.accountname = accountname
        self.date = date        
        self.pay = pay             
        self.shoplist = shoplist
    
    
    # Add transaction record when transaction completed               
    def add_record(self):
        with open('transaction_record.txt', 'a') as trans_file:
            trans_file.write(f'{self.trans_no}                 {self.accountname}     {self.pay/100} dollars       {self.date}       {self.shoplist} \n') 
            
            
    # show item purchased and the purchased amount 
    def show_cart(cart):
        import pandas as pd
        cart_dict = {'Item': [prod.name for prod in cart.keys()],
                    'Quantity': [qtt for qtt in cart.values()]}
        cart_frame = pd.DataFrame(cart_dict)
        return cart_frame 
    
    
    # Calculate total price based on shopping list and quantity for each item                       
    def getprice(shoppingcart):
        total = 0
        for prod in shoppingcart:        
            total += shoppingcart[prod] * prod.cost
        return total


    # This function represents the payment process.  
    def calc_refund(usercredit, totalprice, coinlist):
        import time
        # For each coin inserted, compare total user credit with total price of items. Paymen continues when credit is below payment required.
        while usercredit < totalprice: 
            try:
                # Reminds user of how much more is required.
                print(f'{(totalprice - usercredit) / 100} dollars required.')                
                ctn_ins = input("press any key to insert coins, or 'cancel' to cancel order and get refund: ")
                if ctn_ins == 'cancel':                 
                    totalprice = 0
                    return (-1 - usercredit) 
                else:                                   
                    inserted = int(input('Please insert coin (enter value in cents): '))
                    if inserted in [10, 20, 50, 100, 200]:
                        usercredit += inserted
                        for i in coinlist:
                            if i.val == inserted:
                                i.stor += 1
                        print('Thank you. \n')  
                        continue
                    # System checks coin validity. Reminder message is given when inserted coin is not of acceptable types.
                    else:
                        print("Acceptable coins: 10c, 20c, 50c, $1 (100) or $2 (200). Thank you!")
                        ctn = input("'continue' to insert coin, other key to cancel order and get refund: ")
                        if ctn == 'continue':
                            continue
                        else:                                                                               
                            return (- 1 - usercredit)
                
            except ValueError:
                print('Sorry, numerical value expected.')
                ctn = input("'continue' to insert coin, other key to cancel all: ")
                if ctn == 'continue':
                    continue
                else:                                                            
                    return (-1 - usercredit)
        return usercredit - totalprice

    
    # Store statistical data when required:
    def show_record():
        with open('transaction_record.txt', 'r') as trans_file: 
            for line in trans_file:
                print(line) 
                    
    
            
```

```python
user_dict = {}


class Account:
    def __init__(self, name, credit = 0):        
        self.name = name         
        self.credit = credit  
        
    # Find user by their username, and return their account. 
    # In the purchase process, this function is called to check if user had any previous credit
    def getuser():
        # User set up account or sign in to access previous credits       
        tp_name = input(" If you had an account, enter username: \n Or, enter username to set up new account: ")
        
        for existinguser in user_dict:
            if tp_name == existinguser.name:
                return existinguser
        
        newaccount = Account(tp_name)                     
        return newaccount    

    
```

```python
# Add products: 
coke = Products(1, 'Coke', 250)
coffee = Products(2, 'Coffee', 340)
tea = Products(3, 'Tea', 310, 10)
juice = Products(4, 'Juice', 340)
oat_bar = Products(5, 'Oat Bar', 480)
lolipop = Products(6, 'Lolipop', 160, 0)
prod_list = [coke, coffee, tea, juice, oat_bar, lolipop]
```

```python
# Add coin types
tenc = Coin('10c', 10)
twenc = Coin('20c', 20)
fifc = Coin('50c', 50, 45)
oned = Coin('$1', 100)
twod = Coin('$2', 200)
coin_stora = [tenc, twenc, fifc, oned, twod]
```

```python
def main():
    import datetime     
    
    # Write transaction record headers
    with open('transaction_record.txt', 'w') as trans_file: 
        trans_file.write("Transaction no.    User        Date          Payment         Items purchased               \n")
    user_dict = {}
    
    # Greets customer and prompt to set admin password
    greet()             
    skipline()
    adm_password = Admin.set_pswd()
    admin = True
    
    while admin:
        # Show product and coin storage:
        Products.show_avail(prod_list)                     
        print("\nStored coins:")
        Coin.show(coin_stora)   
        adm_func() 
        adm_chs = input('Please enter: ')
        # First intallation: supplier can reset quantity of products available or reset number of coin stored.            
        if adm_chs.lower() == 'a':            
            Products.res_quan(prod_list)            
        elif adm_chs.lower() == 'b':
            Coin.reset(coin_stora)          
        # Enter user system
        elif adm_chs.lower() == 'c': 
            admin = False
            user_sys = True
            break
        # Exception handling: when choice does not match any function, warning is given.
        else:
            print("Sorry, choice not recognised. Please enter a, b or c.")            
            continue 
            
    # Enter user system
    # Initialise variables
    skipline()
    sugar_patch = 5 
    trans_no = 0
    record_dict = {}      
    user_sys = True 
    
    while user_sys:
        greet() 
        print('System working.') 
        Products.show_avail(prod_list)
        user_func()         
        choice = input('Please enter: ')               
        # Function : check transaction record
        if choice.lower() == 'b':
            Transaction.show_record()
            back = input('Press anywhere to go back to home screen: ') 
            continue
        # Admin function: show error message
        elif choice.lower() == 'c': 
            Admin.err_msg(adm_password)                                      
        # Exception handling
        elif choice.lower() == 'a':                
            # Purchase products:
            skipline()                                                      
            # User selects product
            purchase = True
            cart = {}
            while purchase:
                # User chooses item. If user chose to reset item, function returns -1; if order is cancelled, function returns 0
                if not sugar_patch > 0:
                    print('Please note: sugar is not available.')
                # If user chooses coffee and sugar patch storage is currently 0, alert message will be given again. 
                itmchs = Products.choose_item(prod_list, sugar_patch)
                # Cancel products takes user back to home screen
                if itmchs == 'cancel':
                    purchase = False
                    break
                elif itmchs == 'restart':
                    continue
                else:                    
                    # User selects quantity of items:
                    how_many = Products.set_qtt(itmchs)
                    if how_many == 0:
                        continue
                    elif how_many == -1:
                        purchase = False
                        break
                    else:
                        cart.update({itmchs: how_many}) 
                        skipline()
                        # Shows current items selected and confirm whether user wants to add product 
                        print(Transaction.show_cart(cart))
                        print(f'Total price: {Transaction.getprice(cart) / 100} dollars.')
                        ano_prod = input("'continue to buy' to purchase another item, any other key to start payment: ")
                        if ano_prod.lower() == 'continue to buy':
                            continue                                       
                        else:
                            end_cart = cart                                                                             
                            if end_cart == None:
                                purchase = False
                                continue
                            else:
                                refund = 0 
                                end_price = Transaction.getprice(end_cart)                   

                                # Get user previous credit(if any)
                                thisuser = Account.getuser()
                                if thisuser not in user_dict:
                                    user_dict.update({thisuser: [thisuser.name, thisuser.credit]})

                                credit = thisuser.credit                                
                                print(f'Current credit: {credit / 100} dollars')
                                # Calculate refund:
                                if credit >= end_price:
                                    refund = credit - end_price
                                else:                    
                                    refund = Transaction.calc_refund(credit, end_price, coin_stora)
                                    # When stored coins cannot make exact amount of refund, the remainder is stored in user's credit.
                                if refund < 0 :                              
                                    refund = - (refund + 1)
                                else:                   

                                    # Prepare order. If cancelled, pass                   
                                    if coffee in end_cart:
                                        cup = end_cart[coffee]                                                                                                             
                                        sugar_patch = Products.prepcoffee(sugar_patch, cup) 
                                        print("Thank you. Please wait 15 seconds while the machine prepares other items.")
                                    else: 
                                        print("Thank you. Please wait 15 seconds while the machine prepares items.")                  

                                    # Dispense item
                                    print(f"Please collect: {Transaction.show_cart(end_cart)}") # Dispense items and give refund.                 
                                    for prod in end_cart:
                                        Products.dispense(prod, end_cart[prod]) 
                                    for prod in prod_list:
                                        if prod.name in end_cart.keys():
                                            prod.availty -= end_cart[prod.name]
                                    bought = [] 
                                    for prod in end_cart.keys():
                                        bought.append(prod.name)
                                    trans_no += 1
                                    crt_trans = Transaction(trans_no, thisuser.name, datetime.date.today(), end_price, bought)                                    
                                    Transaction.add_record(crt_trans)

                                # Give refund:                                
                                print(f'Current refund: {refund / 100} dollars')                                                                         
                                thisuser.credit = Coin.get_ref(refund, coin_stora)
                                print(f'Please collect refund {(refund - thisuser.credit)/100} dollars. {thisuser.credit/100} dollars saved to account credit.')      
                                # Goodbye message to user, back to home screen
                                bye()                              
                                back = input('Press anywhere to go back to home screen: ') 
                                purchase = False
                                continue
        else:
            print('Sorry, option not available. Please enter a, b or c: ')
            continue 
            

if __name__ == "__main__":
    main()
    
```
